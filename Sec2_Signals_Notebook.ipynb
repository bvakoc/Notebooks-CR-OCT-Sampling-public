<b>Interactive tools</b>
<ul>
    <li>'PulseWidth' &nbsp;&nbsp;&nbsp;&nbsp;Changes the pulse width. It's FWHM is normalized to the (inverse) pulse repetition rate.</li>
    <li>'$f/f_o$' &nbsp;&nbsp;&nbsp;&nbsp;Changes the RF fringe frequency (signal depth). It is normalized to the optical Nyquist frequency.</li>
    <li>'I/Q' &nbsp;&nbsp;&nbsp;&nbsp;Turns quadrature signal on and off.</li>
    <li>'BG' &nbsp;&nbsp;&nbsp;&nbsp;Tuns background subtraction on and off.</li>
    <li>'Log' &nbsp;&nbsp;&nbsp;&nbsp;Switches between log and linear scale.</li>
    <li>'Norm' &nbsp;&nbsp;&nbsp;&nbsp;Normalizes the signal amplitude. Saves the last normalization value when disabling.</li>
</ul>
