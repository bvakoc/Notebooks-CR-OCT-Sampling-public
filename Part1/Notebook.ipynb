<h1>RF Properties of Circular-Ranging OCT Signals</h1>

This Notebook illustrates the RF frequency space of CR-OCT signals as produced by a monotonically stepped frequency comb source. The output of this simulation represents the signal before low-pass filtering and sampling, i.e., as measured by a RF spectrum analyser. For simplicity, this simulation does not take into account pulse overlap. 

<h3>Description</h3>

<b>Interactive tools</b>

Depicted below is a oversimplified CR-OCT system made of a frequency comb source, interferometer and quadrature signal generation. The system can be controlled by modifying the FC source pulse width, optical delay as well as toggeling I/Q and background substruction using widgets.

| Widget name | Description|
| :---: | :--- |
| PulseWidth | Changes the pulse width (FWHM). Pulse FWHM is normalized to the pulse spacing, $t_p$.|
| $\tau 2\Delta\nu$ | Delay ($\tau$), i.e., signal depth, normalized to the circular delay range, $(2\Delta\nu)^{-1}$. |
| I/Q | Quadrature signal on and off. |
| BG | Background subtraction on and off. |
| Log | Switches between log and linear scale. |

<b>Signal output</b>
<ul type="square">
    <li>The main graph shows the RF frequency space of a CR-OCT signal.</li>
    <li>The x-axis represents the normalized RF frequency with respect to the optical Nyquist frequency, $F_{oN}$.</li>
    <li>The blue shaded area indicates the baseband, ranging $\pm(2t_p)^{-1}=\pm F_{oN}$, corresponding to the circular delay range ($\pm(2\Delta\nu)^{-1}$).</li>
    <li>The brown, dotted curve depicts the signal envelope, $G(f)$.</li>
    <li>The small inset graph shows a pulse shape function, $g(t)$.</li>
</ul>
